```python
top_100_cities_with_state = top100_accidents_by_city.merge(city_state, on='City', how='right')
top_100_cities_with_state = top_100_cities_with_state.dropna(subset=['accidents_by_city'])
top_100_cities_with_state=top_100_cities_with_state.drop_duplicates(subset=['City','accidents_by_city'])
top_100_by_state=top_100_cities_with_state.groupby('State').size().reset_index(name='count').sort_values(by='count',ascending=False)
top_100_by_state
```

## Summarize the inferences & write a conclusion

### Insights

##### - No data from New York
##### - Less than 1% of cities have more than 10000 accidents
##### - Most of the cities have between 0 and 1000 accidents
##### - A high % of accidents occur between 6 am to 10 am , probably people in a hurry to get to work.
##### - Next highest % of accidents is between 3 pm to 6 pm (time to returning from work).
##### - The number of accidents is lower on week-ends.
##### - On weekends, the peak occurs between 10 AM and 4 PM, unlike on weekdays.
##### - The highest percentage of accidents occurs in the winter months.
##### - The number of accidents increasing year over year.
##### - The data is concetrated in some regions
##### - The most of accidents occure when the temperature is between 60 F & 80 F (warmer areas).


### Conclusion

##### *Geographical Distribution:

###### -The dataset does not include any data from New York.
###### -Less than 1% of cities have experienced more than 10,000 accidents.
###### -Most cities have between 0 and 1,000 accidents, indicating a wide distribution across many locations.

##### *Temporal Patterns:

###### -A significant percentage of accidents occur between 6 AM and 10 AM, likely due to people hurrying to get to work.
###### -The next highest percentage of accidents is between 3 PM and 6 PM, corresponding to the time when people are returning from -work.
###### -The number of accidents decreases on weekends, with the peak occurring between 10 AM and 4 PM, unlike on weekdays.

##### *Seasonal Variations:

###### -The highest percentage of accidents occurs during the winter months.
###### -There is a year-over-year increase in the number of accidents, suggesting a rising trend.

###### *Temperature Influence:

###### -Most accidents occur when the temperature is between 60Â°F and 80Â°F, indicating a higher frequency in warmer areas.
